```python
rev_m = 98781.5
rev_stdev = 12118.22
iterations = 1000

rev = np.random.normal(rev_m, rev_stdev, iterations)
CA = - (rev * np.random.normal(0.85,0.16))

CA.mean(), CA.std()
```

```python
Profit = rev + CA
Profit
```

```python
plt.figure(figsize=(15, 6))
plt.plot(Profit)
plt.show()
```

```python
max(Profit)
```

```python
min(Profit)
```
